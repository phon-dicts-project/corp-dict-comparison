```python
import pandas as pd
```

```python
full_df = pd.read_csv('data/andic_dicts.csv')
```

```python
andic_df = full_df.loc[full_df['glottocode'] == 'andi1255']
andic_df.rename(columns={'lemma ': 'lemma'}, inplace=True)
```

```python
andic_df = andic_df.loc[(andic_df['bor'] != 1) & (andic_df['id_meaning'] == 1)]
```

```python
df = andic_df[['id_word', 'id_meaning', 'lemma', 'ipa', 'bor']]
```

```python
def ipaing(text):
    text = text.replace('.', '')
    text = text.replace(',', '')
    text = text.replace('(', '')
    text = text.replace(')', '')
    text = text.replace(' ', '')
    
    text = text.replace('ӏ', '1')
    text = text.replace('кгъ', '-ɢ')
    
    text = text.replace('а', '-a')
    
    text = text.replace('б', '-b')
    
    text = text.replace('в', '-w')
    
    text = text.replace('гъгъ', '-ɢ')
    text = text.replace('гъ', '-ʁ')
    text = text.replace('гi', '-ʕ')
    text = text.replace('г1', '-ʕ')
    text = text.replace('гь', '-h')
    text = text.replace('г', '-ɡ')
    
    text = text.replace('д', '-d')
    
    text = text.replace('е', '-e')
    
    text = text.replace('ё', '-jo')
    
    text = text.replace('ж', '-ʒ')
    
    text = text.replace('з', '-z')
    
    text = text.replace('и', '-i')
    
    text = text.replace('й', '-j')
    
    text = text.replace('к1к1', "-k'ː")
    text = text.replace('кiкi', "-k'ː")
    text = text.replace('къкъ', "-qχ'ː")
    text = text.replace('кькь', "-tɬ'ː")
    text = text.replace('кк', '-kː')
    text = text.replace('кi', "-k'")
    text = text.replace('к1', "-k'")
    text = text.replace('кь', "-tɬ'")
    text = text.replace('къ', "-qχ'")
    text = text.replace('к', '-k')
    
    text = text.replace('л1л1', '-tɬː')
    text = text.replace('лiлi', '-tɬː')
    text = text.replace('лълъ', '-ɬː')
    text = text.replace('л1', '-tɬ')
    text = text.replace('лi', '-tɬ')
    text = text.replace('лъ', '-ɬ')
    text = text.replace('л', '-l')
    
    text = text.replace('м', '-m')
    
    text = text.replace('н', '-n')
    
    text = text.replace('о', '-o')
    
    text = text.replace('п1п1', "-p'ː")
    text = text.replace('пiпi', "-p'ː")
    text = text.replace('пп', '-pː')
    text = text.replace('п1', "-p'")
    text = text.replace('пi', "-p'")
    text = text.replace('п', '-p')
    
    text = text.replace('р', '-r')
    
    text = text.replace('сс', '-sː')
    text = text.replace('с', '-s')
    
    text = text.replace('т1т1', "-t'ː")
    text = text.replace('тiтi', "-t'ː")
    text = text.replace('тт', '-tː')
    text = text.replace('т1', "-t'")
    text = text.replace('тi', "-t'")
    text = text.replace('т', '-t')
    
    text = text.replace('у', '-u')
    
    text = text.replace('ф', '-f')
    
    text = text.replace('хьхь', '-xː')
    text = text.replace('хx', '-χː')
    text = text.replace('хь', '-x')
    text = text.replace('хъ', '-qχ')
    text = text.replace('хi', '-ħ')
    text = text.replace('х1', '-ħ')
    text = text.replace('х', '-χ')
    
    text = text.replace('ц1ц1', "-ts'ː")
    text = text.replace('цiцi', "-ts'ː")
    text = text.replace('цц', '-tsː')
    text = text.replace('ц1', "-ts'")
    text = text.replace('цi', "-ts'")
    text = text.replace('ц', '-ts')
    
    text = text.replace('ч1ч1', "-tʃ'ː")
    text = text.replace('чiчi', "-tʃ'ː")
    text = text.replace('чч', '-tʃː')
    text = text.replace('ч1', "-tʃ'")
    text = text.replace('чi', "-tʃ'")
    text = text.replace('ч', '-tʃ')
    
    text = text.replace('шш', '-ʃː')
    text = text.replace('ш', '-ʃ')
    
    text = text.replace('щ', '-ʃː')
    
    text = text.replace('ъ', '-ʔ')

    text = text.replace('ы', '-i')
    
    text = text.replace('э', '-e')
    text = text.replace('ю', '-j-u')
    text = text.replace('я', '-j-a')
    
    text = text.replace("-k-w", "-kʷ")
    text = text.replace("-k'-w", "-k'ʷ")
    text = text.replace("-ɡ-w", "-ɡʷ")
    text = text.replace("-h-w", "-hʷ")
    text = text.replace("-ʁ-w", "-ʁʷ")
    text = text.replace("-t͡ɬ-w", "-t͡ɬʷ")
    text = text.replace("-q͡χ-w", "-q͡χʷ")
    text = text.replace("-q͡χ'-w", "-q͡χ'ʷ")
    text = text.replace("-t͡ɬ'-w", "-t͡ɬ'ʷ")
    text = text.replace("-t͡ʃ'-w", "-t͡ʃ'ʷ")
    text = text.replace("-t'-w", "-t'ʷ")
    text = text.replace("-χː-w", "-χːʷ")
    text = text.replace("-t͡ʃ-w", "-t͡ʃʷ")
    text = text.replace("-ʃː-w", "-ʃːʷ")
    text = text.replace("-x-w", "-xʷ")
    text = text.replace("-ʃː'-w", "-ʃː'ʷ")
    text = text.replace("-d-w", "-dʷ")
    text = text.replace("-ʒ-w", "-ʒʷ")
    text = text.replace("-s-w", "-sʷ")
    text = text.replace("-sː-w", "-sːʷ")
    text = text.replace("-kː'-w", "-kː'ʷ")
    text = text.replace("-sː'-w", "-sː'ʷ")
    text = text.replace("-χ-w", "-χʷ")
    text = text.replace("-z-w", "-zʷ")
    text = text.replace("-r-w", "-rʷ")
    text = text.replace("-t-w", "-tʷ")
    text = text.replace("-l-w", "-lʷ")
    text = text.replace("-n-w", "-nʷ")
    text = text.replace("-t͡s'-w", "-t͡s'ʷ")
    
    text = text.replace('͡', '')
    
    if len(text) > 0 and text[0] == '-':
        text = text[1:]
    if len(text) > 0 and text[len(text)-1] == '-':
        text = text[:1]
    
    return text
```

```python
new_ipa = []

for i in df['lemma']:
    ipa = ipaing(i.lower())
    new_ipa.append(ipa)
```

```python
df['ipa'] = new_ipa
```

```python
df
```

```python
words = new_ipa
```

```python
type_freq = {}
stop = ['', '\n']
for word in words:
    frags = word.split('-')
    for f in frags:
        if f not in stop:
            if f in type_freq.keys():
                type_freq[f] += 1
            else:
                type_freq[f] = 1
```

```python
type_f = []
type_sum = sum(type_freq.values())
for i in type_freq.values():
    type_f.append(i/type_sum)
```

```python
import pandas as pd
df_type = pd.DataFrame()
df_type['segment'] = type_freq.keys()
df_type['dict_value'] = type_freq.values()
df_type['dict_frequency'] = type_f
```
