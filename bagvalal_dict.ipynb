```python
def no_acc(text):
    text = text.replace("-ã'", "-ã")
    text = text.replace("-aː'", "-aː")
    text = text.replace("-a'", "-a")
    text = text.replace("-ẽ'", "-ẽ")
    text = text.replace("-eː'", "-eː")
    text = text.replace("-e'", "-e")
    text = text.replace("-iː'", "-iː")
    text = text.replace("-i'", "-i")
    text = text.replace("-õ'", "-õ")
    text = text.replace("-oː'", "-oː")
    text = text.replace("-o'", "-o")
    text = text.replace("-uː'", "-uː")
    text = text.replace("-u'", "-u")
    text = text.replace("-uː'", "-uː")
    text = text.replace("u'", "u")
    text = text.replace("uː'", "uː")
    text = text.replace("e'", "e")
    text = text.replace("a'", "a")
    text = text.replace("o'", "o")
    text = text.replace("oː'", "oː")
    text = text.replace('a’','a' )
    text = text.replace("ã'", "ã")
    text = text.replace("i'", "i")
    #and a minor fix
    text = text.replace('l̄-ʔ', '-ɬː')
    text = text.replace('͡', '')
    return text
```

```python
new_ipa = []
for i in df['lemma']:
    ipa = ipaing(i)
    new = no_acc(ipa)
    new_ipa.append(new)
```

```python
words = new_ipa
```

```python
type_freq = {}
stop = ['', '\n']
for word in words:
    frags = word.split('-')
    for f in frags:
        if f not in stop:
            if f in type_freq.keys():
                type_freq[f] += 1
            else:
                type_freq[f] = 1
```

```python
type_f = []
type_sum = sum(type_freq.values())
for i in type_freq.values():
    type_f.append(i/type_sum)
```

```python
import pandas as pd
df_type = pd.DataFrame()
df_type['segment'] = type_freq.keys()
df_type['dict_value'] = type_freq.values()
df_type['dict_frequency'] = type_f
```
