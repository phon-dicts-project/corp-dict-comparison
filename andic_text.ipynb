```python
type_freq = {}
stop = ['', '\n']
frags = data.split('-')
for f in frags:
    if f not in stop:
        if f in type_freq.keys():
            type_freq[f] += 1
        else:
            type_freq[f] = 1
```

```python
type_f = []
type_sum = sum(type_freq.values())
for i in type_freq.values():
    type_f.append(i/type_sum)
```

```python
import pandas as pd
df = pd.DataFrame()
df['segment'] = type_freq.keys()
df['text_value'] = type_freq.values()
df['text_frequency'] = type_f
```
